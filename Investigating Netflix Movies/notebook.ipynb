```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Start coding here! Use as many cells as you like
# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")

# Filter the data to get only movies
netflix_subset_movies = netflix_df[netflix_df["type"] == "Movie"]

# Filter the data to get movies that are released in the 1990s
netflix_subset_nineties = netflix_subset_movies[(netflix_subset_movies["release_year"] >= 1990) & (netflix_subset_movies["release_year"] < 2000)]

# Visualize the duration column of the filtered data to find the most frequent movie duration in the 1990s.
plt.hist(netflix_subset_nineties["duration"])
plt.title('Distribution of Movie Durations in the 1990s')
plt.xlabel('Duration (in minutes)')
plt.ylabel('Number of Movies')
plt.show()

# Filter the data to get the action movies in the 1990s
netflix_subset_action = netflix_subset_nineties[netflix_subset_nineties['genre'] == 'Action']

# Count the number of short action movies
short_movie_count = (netflix_subset_action['duration'] < 90).sum()
print("The number of short action movies released in the 1990s is " + str(short_movie_count) + ".")

# Get the most frequent duration.
duration = netflix_subset_nineties["duration"].mode()[0]
print("The most frequent movie duration in the 1990s is " + str(int(duration)) + ".")
```
